```python
import altair as alt
from vega_datasets import data
import pandas as pd
countries = alt.topo_feature(data.world_110m.url, 'countries')
df=pd.read_csv('https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/ufo-scrubbed-geocoded-time-standardized-00.csv')
df=df.rename(columns={'29.8830556': 'latitude', '-97.9411111': 'longitude','10/10/1949 20:30':'time', 'san marcos':'city', 'tx':'state','us':'country', 'cylinder':'shape','45 minutes':'duration', 'This event took place in early fall around 1949-50. It occurred after a Boy Scout meeting in the Baptist Church. The Baptist Church sit':'description'})
df=df.dropna(subset=['latitude','longitude'])
df5000=df.sample(n=5000)
background = alt.Chart(countries).mark_geoshape(
    fill='black',
    stroke='white'
).project(
    "equirectangular"
).properties(
    title = 'UFO Sightings',
    width=900,
    height=500
)
points = alt.Chart(df5000).mark_circle().encode(
    longitude='longitude:Q',
    latitude='latitude:Q',
    size=alt.value(10),
    color=alt.value('lime'),
    tooltip=['city','state','country','time','duration','shape','description','latitude','longitude']
).project(
    "equirectangular"
).properties(
    width=900,
    height=500
)
chart=background+points
chart
```

```python
df1=pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv')
df1_filter=df1[df1['Year Constructed']!=0]
df1_year=df1_filter.groupby('Year Constructed')['Year Constructed'].count().reset_index(name='Number of Buildings')
chart1=alt.Chart(df1_year).mark_circle().encode(
    x='Year Constructed:O',
    y='Number of Buildings:Q',
    color=alt.value('red')
).properties(
    title="# of Buildings Constructed Per Year",
    width=900,
    height=500
)
chart1
```
